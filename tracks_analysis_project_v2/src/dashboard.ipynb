```python
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://user:password@localhost:5432/tracks')
df = pd.read_sql("SELECT * FROM tracks_with_clusters", engine)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
```

```python
axes[0,0].bar(df['cluster'].value_counts().index, 
              df['cluster'].value_counts().values)
axes[0,0].set_title('Распределение треков по кластерам')
axes[0,0].set_xlabel('Кластер')
axes[0,0].set_ylabel('Количество')
```

```python
scatter = axes[0,1].scatter(df['distance_km'], df['elevation_gain'], 
                           c=df['cluster'], cmap='viridis', alpha=0.6)
axes[0,1].set_title('Дистанция vs Набор высоты')
axes[0,1].set_xlabel('Дистанция (км)')
axes[0,1].set_ylabel('Набор высоты (м)')
plt.colorbar(scatter, ax=axes[0,1])
```

```python
cluster_data = [df[df['cluster']==i]['distance_km'] for i in df['cluster'].unique()]
axes[1,0].boxplot(cluster_data)
axes[1,0].set_title('Распределение дистанций по кластерам')
axes[1,0].set_xlabel('Кластер')
axes[1,0].set_ylabel('Дистанция (км)')
```

```python
cluster_means = df.groupby('cluster')[['distance_km', 'elevation_gain']].mean()
cluster_means.plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title('Средние значения по кластерам')
axes[1,1].set_xlabel('Кластер')
```

```python
plt.tight_layout()
plt.savefig('../outputs/clusters_visualization.png', dpi=100)
plt.show()
```
